### 2.2. Изучение распределения этажей домов:

```python
import matplotlib.pyplot as plt

data = kc_house_data['floors'].value_counts()

plt.figure(figsize=(25, 25))
colors = plt.cm.Paired(range(len(data)))
explode = [0.0, 0.1, 0.3, 0.5, 0.5, 0.5]
plt.pie(data, autopct='%1.1f%%', labels=data.index, colors=colors, startangle=140, explode=explode, textprops={'fontsize': 14})
plt.legend(title="Этажи", title_fontsize="13", fontsize="12", bbox_to_anchor=(1, 0.5), loc="center left")
plt.title('Количество этажей', fontsize=16)
plt.show()
```

Выводы по заданию 2.2:
- Большая часть продаваемых домов имеет 1 и 2 этажа.
- Самые продаваемые дома - одноэтажные (49,4%).

### 2.3. Изучите распределение состояния домов

```python
data = kc_house_data['condition'].value_counts()

plt.figure(figsize=(15, 15))
explode = [0.1] * len(data) 
colors = plt.cm.Paired(range(len(data)))

plt.pie(data, autopct='%1.1f%%', labels=data.index, explode=explode, colors=colors, startangle=140, textprops={'fontsize': 14})
plt.title('Состояние домов', fontsize=16)
plt.legend(title="Состояние", title_fontsize="13", fontsize="12", bbox_to_anchor=(1, 0.5), loc="center left")
plt.show()
```

Выводы по заданию 2.3:

- Значительное количество домов имеет состояние выше среднего (оценка 4).
- Домов в отличном состоянии (оценка 5) меньше, чем домов с оценкой 3 или 4.
- Дома в плохом (оценка 1) или ниже среднего (оценка 2) состоянии встречаются редко.

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

1. Корреляционная матрица

```python
corr_matrix = kc_house_data.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Корреляционная матрица характеристик недвижимости')
plt.show()
```

Вывод: 

Корреляционная матрица показывает, что наиболее сильная связь с ценой наблюдается у площади жилья (sqft_living), оценки качества (grade) и количества ванных комнат (bathrooms). Слабые связи (меньше 0.3) были заменены на 0 для упрощения анализа.

2. Зависимость цены от площади жилья

```python
sns.jointplot(x=kc_house_data['sqft_living'], y=kc_house_data['price'], kind='reg')
plt.show()
```

Вывод: 

Наблюдается сильная положительная корреляция между ценой и площадью жилья. С увеличением площади цена имеет тенденцию к росту, хотя разброс увеличивается для больших площадей.

3. Зависимость цены от оценки качества

```python
sns.jointplot(x=kc_house_data['grade'], y=kc_house_data['price'], kind='reg')
plt.show()
```

Вывод: 

Существует положительная корреляция между ценой и оценкой качества дома. Дома с более высокой оценкой качества, как правило, стоят дороже.

4. Распределение цены в зависимости от класса дома

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x=kc_house_data['grade'], y=kc_house_data['price'], whis=1.0)
plt.xlabel('Класс дома')
plt.ylabel('Цена')
plt.title('Распределение цен в зависимости от класса дома')
plt.show()
```

Вывод: 

С повышением класса дома медианная цена растет, также увеличивается разброс цен.

5. Распределение средних цен в зависимости от года постройки

```python
data = kc_house_data.groupby('yr_built')['price'].mean().reset_index()
plt.figure(figsize=(15, 6))
plt.scatter(data['yr_built'], data['price'])
plt.title('Распределение средних цен в зависимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Средняя цена')
plt.show()
```

Вывод: 

Более новые дома в среднем дороже, но зависимость не линейная.

5. Распределение цены по количеству этажей

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=kc_house_data['floors'].astype('str'), y=kc_house_data['price'], whis=1.5)
plt.xlabel('Количество этажей')
plt.ylabel('Цена')
plt.title('Распределение цены по количеству этажей')
plt.show()
```

Вывод: 

Наблюдается тенденция к повышению цены с увеличением количества этажей, хотя эта зависимость не столь выражена, как в случае с площадью или оценкой качества

6. Распределение цены по наличию береговой линии

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x=kc_house_data['price'], y=kc_house_data['waterfront'].astype('str'), whis=1.5, hue=kc_house_data['grade'])
plt.title('Распределение цен по наличию береговой линии')
plt.xlabel('Цена')
plt.ylabel('Наличие набережной')
plt.show()
```

Вывод: 

Дома у воды значительно дороже, особенно высококлассные.
